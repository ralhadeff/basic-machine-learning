# Gaussian mixture model clustering

```python
import matplotlib.pyplot as plt
import numpy as np

from gaussian_mixture import GaussianMixtureClustering
```

### Data

```python
size = 100
# blob 1
x = np.random.normal(-4,1.5,(size))
y = np.random.normal(-12,1.5,(size))
# blob 2
x = np.concatenate((x,np.random.normal(0,1.5,(size))))
y = np.concatenate((y,np.random.normal(1,1.5,(size))))
# blob 3
x = np.concatenate((x,np.random.normal(-7,1.5,(size))))
y = np.concatenate((y,np.random.normal(4,1.5,(size))))
# blob 4
x = np.concatenate((x,np.random.normal(-5,1.5,(size))))
y = np.concatenate((y,np.random.normal(-3,1.5,(size))))

X = np.hstack((x[:,None],y[:,None]))
```

### Fitting and predicting

```python
gmm = GaussianMixtureClustering()
gmm.fit(X,5)
y = gmm.predict(X)
```

### Visualize clusters

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

Note: initial Gaussians result in variation of output. Interstingly, the blue cluster here is a very narrow Gaussian, and therefore the single yellow point below it is considered part of the larger area of the broader yellow Gaussian, rather than the very thin blue one.
