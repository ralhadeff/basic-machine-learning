## matplotlib.pyplot styles

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
y = np.arange(10)**2
z = np.random.normal(size=100)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3,figsize=(16,8))
axes[0][0].scatter(X,y)
axes[0][1].plot(X,y)
axes[0][2].bar([1,2,3,4],[3,7,1,2])
axes[1][0].boxplot(X)
axes[1][1].hist(z)
axes[1][2].pie([3,5,8,2],explode=[0.2,0.05,0.05,0.05])
plt.show()
```

```python
plt.style.use('ggplot')
with plt.xkcd():
    fig, axes = plt.subplots(nrows=2, ncols=3,figsize=(16,8))
    axes[0][0].scatter(X,y)
    axes[0][1].plot(X,y)
    axes[0][2].bar([1,2,3,4],[3,7,1,2])
    axes[1][0].boxplot(X)
    axes[1][1].hist(z)
    axes[1][2].pie([3,5,8,2])
    plt.show()
```
