# Global pooling

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

from conv2d import Conv2D
```

```python
class GlobalPooling():
    
    def pool(self,X,pool='average'):
        if (pool=='ave' or pool=='average' or pool=='mean'):
            return X.mean(axis=0).mean(axis=0)
        else:
            return X.max(axis=0).max(axis=0)
```

### Sample image

```python
image = mpimg.imread('gpcr.png')
# make image smaller for faster execution
image = (image[::6,::6,:3])
```

```python
plt.imshow(image)
plt.show()
```

### Convolution and global pooling

```python
conv = Conv2D(no_of_kernels=3,color_channels=3,kernel_size=3,activation='ReLU')
conv.initialize_weights()
```

```python
convolved = conv.feed_forward(image)
plt.imshow(convolved)
plt.show()
```

```python
convolved.shape
```

```python
# average pooling by default
pool = GlobalPooling()

pool.pool(convolved)
```

```python
conv = Conv2D(no_of_kernels=64,color_channels=3,kernel_size=3,activation='ReLU')
conv.initialize_weights()

convolved = conv.feed_forward(image)
convolved.shape
```

```python
# max pooling specified
pool.pool(convolved,pool='max')
```
