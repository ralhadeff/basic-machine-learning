```python
import numpy as np
```

```python
class Lander():
    
    def __init__(self):
        self.pos = np.array([0.,1000])
        self.vel = np.array([0.,0])
        self.acc = np.array([0.,0])
        self.grav = np.array([0.,-5])
        self.fuel = 1000

    def step(self,dt=1):
        # one timestep
        # for simplicity no leap-frog or similar algorithm
        self.vel+=dt*(self.acc+self.grav)
        self.pos+=dt*self.vel
        
    def check_landing(self):
        # check if landed on the surface
        return (self.pos[1]<=0)
    
    def burner(self,mag):
        if (self.fuel>=mag):
            self.acc = np.array([0.,mag])
            #self.fuel-=mag
        else:
            self.acc = np.array([0.,0])
      
    def state(self):
        return (self.pos[1],self.vel[1])
```

```python
lander = Lander()

for i in range(100000):
    lander.burner(4.88)
    lander.step(0.1)
    if (lander.check_landing()):
        if (lander.vel[1]<-5):
            print('lost')
        else:
            print('success')
        break
```
