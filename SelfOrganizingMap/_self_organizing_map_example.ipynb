# Self organizing map

```python
import numpy as np
import matplotlib.pyplot as plt

from som import SOM
```

### Generate shaped data

```python
size = 500

x = np.random.normal(0,1,size)
y = np.random.normal(-1,1,size)
x[x>0]*=3
y[(x<1.5) * (y<0)]*=3

X = np.concatenate((x,y)).reshape(2,size).T
plt.scatter(X[:,0],X[:,1],s=20)
plt.show()
```

### Generate map

```python
som = SOM(5,5)
som.fit(X,1000)
```

### Visualize nodes

**Note: this is the position of the nodes in the feature space, it is NOT the map. This representation is only possible in 1D, 2D or 3D, whereas the map is ALWAYS 2D (and the feature space can have as many features as the performance allows**

```python
plt.scatter(X[:,0],X[:,1],s=20)
plt.scatter(som.nodes[:,:,0],som.nodes[:,:,1],s=80,c='black')

plt.show()
```

As can be seen, the grid covers the data and its shape

### Visualize map

```python
distance_map = som.get_map()
```

```python
plt.imshow(distance_map)
plt.show()
```

There are no clear clusters in the data to begin with

### Clustered data

```python
size = 500

x = np.concatenate((np.random.normal(0,1,size),np.random.normal(-6,1,size),np.random.normal(5,1,size)))
y = np.concatenate((np.random.normal(0,1,size),np.random.normal(-6,1,size),np.random.normal(-6,1,size)))

X = np.concatenate((x,y)).reshape(2,3*size).T
plt.scatter(X[:,0],X[:,1],s=20)
plt.show()
```

```python
som = SOM(10,10)
som.fit(X,1000)
```

```python
# in samples space
plt.scatter(X[:,0],X[:,1],s=20)
plt.scatter(som.nodes[:,:,0],som.nodes[:,:,1],s=80,c='black')

plt.show()
```

```python
# in grid space
distance_map = som.get_map()
plt.imshow(distance_map)
plt.show()
```

**Three clusters are observed**

### Individual maps

```python
# x axis - map is about the same
distance_map = som.get_map(0)
plt.imshow(distance_map)
plt.show()
```

```python
# y axis - only two clusters are now observed
distance_map = som.get_map(1)
plt.imshow(distance_map)
plt.show()
```
